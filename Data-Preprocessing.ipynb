```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("Placement.csv")
dataset
```

```python
#To know null values
dataset.isnull().sum()
```

```python
#To fill empty column into 0 value
dataset["salary"].fillna(0,inplace=True)
```

```python
dataset
```

#to delete entire row

dataset.dropna(inplace=True)

#to replace the empty column into mean,median & mode values

dataset["salary"].fillna(dataset["salary"].mean(),inplace=True)

```python
def QuanQual(dataset):
        qual=[]
        quan=[]
        for columnName in dataset.columns:
            if(dataset[columnName].dtypes=='O'):
                qual.append(columnName)   
            else:
                quan.append(columnName)
        return qual,quan
```

```python
qual,quan=QuanQual(dataset)
```

```python
#for multi columns & rows nan values replace
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
```

```python
df
```

```python
#To convert df values into dataset table format
df=pd.DataFrame(df,columns = quan)
```
